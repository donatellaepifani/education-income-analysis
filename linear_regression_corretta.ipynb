```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Dati per la prima regressione
RLOSSF = np.array([12.61, 10.48, 10.24, 11.34, 10.88, 10.29, 12.00, 15.02, 10.52, 9.99, 9.46, 9.62, 10.29, 9.90, 8.15])
RLODF = np.array([14.73, 11.82, 11.16, 11.77, 14.38, 10.39, 13.53, 21.26, 11.32, 11.57, 10.38, 10.37, 10.52, 10.44, 8.46])
RLOLF = np.array([16.17, 12.61, 13.03, 13.76, 15.25, 10.75, 15.33, 21.25, 12.22, 13.58, 11.86, 12.36, 11.50, 11.08, 10.53])
y1 = np.concatenate((RLOSSF, RLODF, RLOLF))
x1 = np.concatenate((np.zeros_like(RLOSSF), np.ones_like(RLODF), np.full_like(RLOLF, 2)))

# Prima regressione
X1 = sm.add_constant(x1)
model1 = sm.OLS(y1, X1).fit()

# Dettagli statistici della prima regressione
print("Prima Regressione:")
print(model1.summary())

# Dati per la seconda regressione
RLOSSM = np.array([14.30, 11.56, 10.93, 11.27, 12.09, 10.38, 15.39, 22.57, 11.19, 10.98, 10.15, 10.33, 10.39, 10.24, 9.37])
RLODM = np.array([14.94, 11.68, 11.85, 12.03, 13.28, 10.68, 16.53, 26.38, 11.95, 12.91, 10.66, 10.84, 10.83, 10.82, 10.16])
RLOLM = np.array([17.91, 13.79, 16.10, 16.03, 15.34, 10.97, 17.87, 26.22, 16.17, 17.54, 12.17, 14.15, 13.23, 11.53, 11.66])
y2 = np.concatenate((RLOSSM, RLODM, RLOLM))
x2 = np.concatenate((np.zeros_like(RLOSSM), np.ones_like(RLODM), np.full_like(RLOLM, 2)))

# Seconda regressione
X2 = sm.add_constant(x2)
model2 = sm.OLS(y2, X2).fit()

# Dettagli statistici della seconda regressione
print("\nSeconda Regressione:")
print(model2.summary())

# Grafico delle due rette di regressione
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, color='blue', label='Dati Prima Regressione')
plt.plot(x1, model1.predict(X1), color='blue', label='Rette Prima Regressione')
plt.scatter(x2, y2, color='red', label='Dati Seconda Regressione')
plt.plot(x2, model2.predict(X2), color='red', label='Rette Seconda Regressione')

# Dettagli del grafico
plt.title("Rette di Regressione Lineare")
plt.xlabel("x (0 = Prima Categoria, 1 = Seconda Categoria, 2 = Terza Categoria)")
plt.ylabel("y (Valori)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Dati per la prima regressione (F)
RLOSSF = np.array([12.61, 10.48, 10.24, 11.34, 10.88, 10.29, 12.00, 15.02, 10.52, 9.99, 9.46, 9.62, 10.29, 9.90, 8.15])
RLODF = np.array([14.73, 11.82, 11.16, 11.77, 14.38, 10.39, 13.53, 21.26, 11.32, 11.57, 10.38, 10.37, 10.52, 10.44, 8.46])
RLOLF = np.array([16.17, 12.61, 13.03, 13.76, 15.25, 10.75, 15.33, 21.25, 12.22, 13.58, 11.86, 12.36, 11.50, 11.08, 10.53])
y1 = np.concatenate((RLOSSF, RLODF, RLOLF))
x1 = np.concatenate((np.zeros_like(RLOSSF), np.ones_like(RLODF), np.full_like(RLOLF, 2)))

# Prima regressione
X1 = sm.add_constant(x1)
model1 = sm.OLS(y1, X1).fit()

# Estrazione dei coefficienti e dei p-value per la prima regressione
intercept1 = model1.params[0]
coef1 = model1.params[1]
pval_intercept1 = model1.pvalues[0]
pval_coef1 = model1.pvalues[1]

# Dati per la seconda regressione (M)
RLOSSM = np.array([14.30, 11.56, 10.93, 11.27, 12.09, 10.38, 15.39, 22.57, 11.19, 10.98, 10.15, 10.33, 10.39, 10.24, 9.37])
RLODM = np.array([14.94, 11.68, 11.85, 12.03, 13.28, 10.68, 16.53, 26.38, 11.95, 12.91, 10.66, 10.84, 10.83, 10.82, 10.16])
RLOLM = np.array([17.91, 13.79, 16.10, 16.03, 15.34, 10.97, 17.87, 26.22, 16.17, 17.54, 12.17, 14.15, 13.23, 11.53, 11.66])
y2 = np.concatenate((RLOSSM, RLODM, RLOLM))
x2 = np.concatenate((np.zeros_like(RLOSSM), np.ones_like(RLODM), np.full_like(RLOLM, 2)))

# Seconda regressione
X2 = sm.add_constant(x2)
model2 = sm.OLS(y2, X2).fit()

# Estrazione dei coefficienti e dei p-value per la seconda regressione
intercept2 = model2.params[0]
coef2 = model2.params[1]
pval_intercept2 = model2.pvalues[0]
pval_coef2 = model2.pvalues[1]

# Output statistici
print("Prima Regressione (F):")
print(model1.summary())
print(f"Intercetta: {intercept1:.4f}, p-value: {pval_intercept1:.4e}")
print(f"Coefficiente: {coef1:.4f}, p-value: {pval_coef1:.4e}")

print("\nSeconda Regressione (M):")
print(model2.summary())
print(f"Intercetta: {intercept2:.4f}, p-value: {pval_intercept2:.4e}")
print(f"Coefficiente: {coef2:.4f}, p-value: {pval_coef2:.4e}")

# Grafico delle due rette di regressione
plt.figure(figsize=(10, 6))

# Dati e retta per la prima regressione (F)
plt.scatter(x1, y1, color='blue', label='Dati F')
plt.plot(x1, model1.predict(X1), color='blue', label='Retta di regressione femminile')

# Dati e retta per la seconda regressione (M)
plt.scatter(x2, y2, color='red', label='Dati M')
plt.plot(x2, model2.predict(X2), color='red', label='Retta di regressione maschile')

# Personalizzazione dell'asse x
plt.xticks([0, 1, 2], ['RLOSS', 'RLOD', 'RLOL'])
plt.xlabel("Livello di studio")
plt.ylabel("Retribuzione oraria lorda ()")
plt.title("Rette di Regressione Lineare per Generi e Livelli di Studio")
plt.legend()
plt.grid(True)

plt.show()
```
